이 글은 확률과정론 시험을 보고 충격받아서 **다시는 이런 실수를 반복하지 않기위해 쓰는 글**이다. 언젠가 또 이 글을 볼 미래의 나에게 말해주자면 만약 이러한 것들 중 하나라도 **딱 하나라도 지금 너가 하고 있다면 그 시험(시험,발표 등등 고난)은 매우 위험하며 결코 성공하지 못할 것**이다. **그 순간 넌 이미 자신에게 지나치게 관대한 인간**이며 **열심히 하는 척 연기하는 사람**이다.

# <span style ="color:black">**절대로 다 안다고 생각하지 말기(자만,가장 위험한 생각)**

- 아마 공부를 좀 해봤을 수도 있다.
- 그래서 "이 정도면 열심히 했다.","이 정도면 다 알지","그건 좀 쉽잖아" 라는 생각이 들 수 있다.
- 가장 위험한 생각이다.
- 이런 생각은 내 마음속에 들어가있는 순간부터 **앞으로 매 순간 최선을 다 하지 못하게** 만들기 때문이다.
- 제일 버려야 할 생각이며 가장 쓸모없다.
- 대충,어느정도만 하게 만들며 계속해서 쌓이다가 결국 나중에 큰 문제가 되어 터지고 만다.
- 또한 가장 가소로운 생각이기도 하다.
- "그건 좀 쉽잖아?" 내가 뭘 안다고 그걸 평가하는가? 내가 뭘 평가할 수준은 되는가?
- "더 이상 할거 없는데"라고 생각이 들 수 있다. 그럴때는 이번에 완전 처음 공부해 본 개념 or 많이 막혔던 내용 or 바로 대답 못하고 머릿속을 한 번 거쳐야 하는 내용 or 밑바닥이 되는 내용 or 문제 만들어서 풀기 등등... 모든 수단을 가동해서 더 공부해야 할 부분을 찾는다,

# <span style ="color:black">**시험기간은 원래 지옥이다.**

- 시험기간은 모두에게 다 힘든 시간이다. 근데 자꾸 벗어나서 편하게 하려하지 말자.
- 시험기간인데 편하고 싶어? 그럼 시험 끝나고가 지옥일 거야...
- 그냥 미리 맞는거라 생각하자. 그래도 성적 잘 맞고 지옥인게 좋잖아?

# <span style ="color:black">**허용된 부분하에서 최대한 쉽게 시험 만들기**

- 뭐든지 간에 어느정도 까지 허용된 부분이 있을 것이다.(확률과정론의 경우 내가 20분을 소요했던 1,2번이 그냥 집에서 적어올 수 있는 문제였다.)
- 허용된 부분을 최대한 활용하자. 최대한 쉽게 시험을 쉽게 만들자.
- 어디까찌 허용이 되는지 세세히 알기 위해서라도 **평소에 열심히 듣는 것이 중요하다. 자만은 금물이며 시간을 헛되이 보내지 말자**

# <span style = "color:black"> **몇 번씩 봤는지 체크하기**

- 어디를 몇 번씩 보며 체크해야 할지가 좀 추상적일 수 있다.
- 시험이라면 봐야할 범위에 있는 여러 단원,부분,문제가 있고 발표준비라면 ppt의 여러 슬라이드가 될 수 있고 등등 뭐냐에 따라서 여러가지 섹션별로 나눠진다.
- 그 중 **가장 최소가 되는 단위**로 나눠서 반드시 체크한다. 
- 이렇게 해야 **어디를 많이 봤고 어디를 적게 봤는지 알 수 있다.**
- 또한 **잘 했던 것과 못했던 것 나눠서 다른 모양으로 체크하자. 이렇게 하면 어디가 많이 막혔는지,잘 안받아 들여지는지 알 수 있다.**

# <span style = "color:black"> **3~4일전 무조건 전자기기 멀리하기**

- 뇌를 공부이외의 다른 것들로부터 비우는 것이 목적이다.
- 시험때 자꾸 노래 생각나서 망함(HOT 캔디 확률과정론 시험 때 자꾸 생각나서 멸망...)
- 유튜브가 문제인 듯 함. 유튜브에서 노래가 우연히 나와가지고 좋아서 계속들음
- 그냥 아예 전자기기들 자체를 멀리 하자.
- 쓸 수 있는 경우는 강의 들을 때이거나 연락온 것 체크해야 할때(이것도 반드시 시간을 정해놓자)
- 사실 3~4일 이라고 해놨지만 중요하게 급히 처리해야 할 일이 있을때는 그냥 멀리하는게 답이다.

# <span style = "color:black"> **평소에 가장 시야에 뭔가 없는 곳에 앉자**

- 뒷 자리나 사람들이 있는 곳에 앉지마라. 앞 자리에 앉아라
- 집중력을 분산시키고 특히 시험때 다른사람이 빨리 풀거나 하는 것을 보면 조급해지기 때문이다.
- 최대한 아무도 없는 온전히 집중할 수 있는 곳을 찾아서 그 장소에서 풀자.(잘 푸는 사람 옆이나 뒤는 특히 반드시 피하자.)
