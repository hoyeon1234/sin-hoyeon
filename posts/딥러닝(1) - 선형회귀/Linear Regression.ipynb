# 선형회귀란?
확률변수사이의 관계가 **선형**이라고 가정했을 때,주어진 표본에 가장 잘 맞는 가중치를 찾는 것입니다.<br> 주어진 표본에 가장 잘 맞는 직선을 찾으면 우리는 두 변수사이에 선형관계가 이렇다라고 추정할 수 있습니다.<br><br>

확률변수사이의 관계가 다음과 같다고 가정해봅시다.<br>
확률변수는 사이의 **관계는 선형**이나 무시할만한 **아주 작은 오차**를 포함합니다.<br>
$Y = w_0 + w_1X_1  ... + w_nX_m + \epsilon$<br><br>

위와 같은 관계를 가지는 확률변수로부터 표본추출(sampling)하여 n개의 표본을 얻으면 다음과 같습니다.<br>
$y_1 = w_0 + w_1x_{11} + ... + w_nx_{1m} + \epsilon_1$<br>
$y_2 = w_0 + w_1x_{21} + ... + w_nx_{2m} + \epsilon_2$<br>
...<br>
$y_n = w_0 + w_1x_{n1} + w_nx_{nm} + \epsilon_n$<br><br>

샘플을 벡터-행렬로 표현해서 좀 더 간단히 나타낼 수 있습니다.<br>
$\begin{bmatrix} y_1 \\ y_2 \\ ... \\y_n\end{bmatrix} = $
$\begin{bmatrix} 
1 & x_{11} & x_{12} & ... &x_{1m}\\
1 & x_{21} & x_{22} & ... &x_{2m}\\
1 & x_{31} & x_{32} & ... &x_{3m}\\
\end{bmatrix}$
$
\begin{bmatrix}
w_0\\
w_1\\
w_2\\
...\\
w_m
\end{bmatrix}
$
$+
\begin{bmatrix}
\epsilon_0\\
\epsilon_1\\
\epsilon_2\\
...\\
\epsilon_n
\end{bmatrix}
$<br>
$\Longleftrightarrow$
$\bold{y} = \bold{X}\bold{W} + \bold{\epsilon}$<br><br>

선형회귀의 목적은 위와 같이 주어진 샘플이 주어질 때, 가장 잘 맞는 $\hat{\bold{W}}$을 추정하는 것입니다.<br>
표본으로부터 이를 구하여 두 변수사이의 관계가 이렇겠구나라고 알고 싶습니다.<br>
