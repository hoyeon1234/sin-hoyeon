# 3 숫자 맞추기 게임
- 변수명 잘 지정하기 ..
- num,answer로 변수명을 지정했었음
    - answer가 컴퓨터가 가지고 있는 정답으로 헷갈려서 조건문 잘못 작성

```python
import random
print("1부터 100 사이의 숫자를 맞추세요")
num = random.randint(1,100)
guess = 0
count = 0
while guess != num:
    guess = int(input("숫자를 입력하시오"))
    if num > guess:
        print("높음")
    elif num < guess:
        print("낮음")
    count +=1
print("축하합니다. 시도횟수=",count)
```

# 암호해독

```python
import random
import sys
user_password = input("암호를 입력하시오 : ")
print("입력된 암호 :",user_password)
password = list("abcdefghijklmnopqurstuvwxyz")
for c1 in password:
    for c2 in password:
        for c3 in password:
            chrs = c1+c2+c3
            print(chrs)
            if chrs == user_password:
                print("찾아낸 암호 :",chrs)
                sys.exit()
```

# 모음의 갯수 카운팅

```python
def count_vowel(string):
    _t = ["a","i","o","u","e"]
    count = 0
    for chr in string:
        if chr in _t:
            count+=1
    return count
```

```python
s = input("문자열을 입력해주세요")
n = count_vowel(s)
print("모음의 갯수는 {}입니다.".format(n))
```

# 변수의 범위

```python
def calculate_area(radius):
    result = 3.14 * radius ** 2
    return result
```

```python
r = 0.5
area = calculate_area(r)
print(result)
```

- result는 함수내부에서 생성된 지역변수이다.
- 지역변수는 함수가 종료되면 사라지게 된다.
- 따라서 함수밖에서 실행하면 오류가 발생하게 된다.

```python
r,area
```

```python
def calculate_area():
    result = 3.14 * r ** 2
    return result
```

```python
r = float(input("원의 반지름: "))
area = calculate_area()
```

```python
print(r,area)
```

- 함수가 무리없이 실행된다.
- 그 이유는 r,area는 외부에서 생성된 변수로 프로그램 어디서나 사용할 수 있기 때문이다.

```python
area = 0
print(id(area))
def calculate_area(radius):
    area = 3.14 * radius **2
    print(id(area))
    return area
r = 5
calculate_area(r)
print(area)
```

- 함수 안에서 area의 값을 바꾸려해도 안바뀐다.
- 함수바깥에서 생성된 전역변수 area외에 함수내부에서 지역변수 area를 따로 생성하기 때문이다.
- 그럼 함수 안에서 전역변수의 값을 바꾸려면 어떻게 해야할까?
    - global로 전역변수라고 선언해주자.

```python
area = 0
print(id(area))
def calculate_area(radius):
    global area
    area = 3.14 * radius **2
    print(id(area))
    return area
r = 5
calculate_area(r)
print(area)
```

- area변수에 있는 값이 바뀌었음을 확인
- 사실 파이썬에서 변수의 값을 바꾼다는 것은?
    - 아예 다른주소를 참조하도록 만드는 것이다.(id확인, 참조하는 위치가 아예 바뀌었다.)

```python
total = 0
print(id(total))
total = 20
print(id(total))
total = 3
print(id(total))
total = 5
print(id(total))
```

- 변수의 값 바꾸기
- 서로다른 메모리주소에 있는 객체 참조.

```python
total = 0
print(id(total))
def update():
    global total
    total = total + 3
update()
print(id(total))
print(total)
```

- global은 함수내부에서 전역변수를 생성 or 변경하기 위해 미리 선언하는 역할이다.
    - 함수내부에서 전역변수를 생성할 수 있다
    - 동일한 이름의 전역변수가 이미 있다면 값을 변경할 수 있다.

# 3개중에서 가장 큰 값 찾기

```python
def get_max(a,b,c=-10000):
    if (a >= b) and (a>=c):
        answer = a
    elif (b >= a) and (b>=c):
        answer =b
    else:
        answer = c
    return answer
```

```python
get_max(10,15)
```

- 위치 인수는 키워드 인수 앞에 나와야 한다.

# 영한사전

```python
english_dict = {}
english_dict["one"] = "하나"
english_dict["two"] = "둘"
english_dict["three"] = "셋"

while True:
    word = input("단어를 입력하시오(종료는 q): ")
    # if문 위치 조심
    # print문 뒤에 오면 q라는 단어자체가 key에 없어서 오류 발생
    if word == "q":
        break
    print(english_dict[word])
    
```

# 일정애플리케이션

```python
task = {}
while True:
    day = input("날짜를 입력")
    if day == "q":
        break
    job = input("일정을 입력")
    task[day] = job
print(task)
```
