```python
from torch.distributions.multivariate_normal import MultivariateNormal
import torch
import matplotlib.pyplot as plt
```

```python
#다변량 정규분포 정의
torch.cuda.manual_seed(26)
mean = torch.tensor([4,-5]).float()
covariance = torch.tensor([3,2,2,3]).reshape(2,2).float()
m = MultivariateNormal(loc=mean,covariance_matrix=covariance)

#샘플링
n_sampling = 50
sample = torch.tensor([m.sample().tolist() for i in range(n_sampling)])
coord_x = sample[:,0]
coord_y = sample[:,1]
x_mean = coord_x.mean().tolist()
y_mean = coord_y.mean().tolist()

#plotting
plt.scatter(coord_x,coord_y,color="green",alpha=0.5)
plt.xlim(-11,11)
plt.ylim(-10,10)
plt.axvline(0,color="black",linestyle="--")
plt.axhline(0,color="black",linestyle="--")
plt.show()
```

```python
plt.arrow(x_mean,y_mean,4,4,head_width=0.7,head_length=0.7,fc="k",ec="k",linewidth=3)
#plotting
plt.scatter(coord_x,coord_y,color="green",alpha=0.5)
plt.xlim(-11,11)
plt.ylim(-10,10)
plt.axvline(0,color="black",linestyle="--")
plt.axhline(0,color="black",linestyle="--")
plt.show()
```

```python
#plotting
plt.arrow(0,0,4,4,head_width=0.7,head_length=0.7,fc="k",ec="k",linewidth=3)
plt.scatter(coord_x-x_mean,coord_y-y_mean,color="green",alpha=0.5)
plt.xlim(-11,11)
plt.ylim(-10,10)
plt.axvline(0,color="black",linestyle="--")
plt.axhline(0,color="black",linestyle="--")
plt.show()
```
